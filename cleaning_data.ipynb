```python
import numpy as np
import pandas as pd
```

```python
# Reading CASE-SHILLER Index into a dataframe
df_CS = pd.read_csv("raw_data/case_shiller_index.csv")

# Selecting data from 2004-01-01 onwards
mask = df_CS["DATE"] >= "2004-01-01"
df_CS = df_CS[mask]

# Resetting Index
df_CS.reset_index(inplace=True, drop=True)

#Understand shape of the data
df_CS.shape

df_CS.head()
```

```python
df_CS.to_csv("cleaned_data/case_shiller_index.csv")
```
